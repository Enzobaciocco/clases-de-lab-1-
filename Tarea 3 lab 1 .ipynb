```python
import numpy as np
import matplotlib.pyplot as plt
import math
from scipy.stats import linregress
```

Para los datos obtenidos del medidor sobre la mesa

```python
#IMPORTAMOS LOS DATOS
datos = np.genfromtxt('datos1.txt',delimiter=",")
y = np.array(datos[:,3])
```

```python
#FILTROS
ylista = list(y)

yfiltro1 = [x for x in ylista if math.isnan(x) == False]          #ELIMINAMOS LOS VALORES QUE NO SON NUMEROS 

yfiltro2 = yfiltro1[5000:]                                        #ELIMINAMOS LOS PRIMEROS CINCO MIL ELEMENTOS 
len(yfiltro2)

yfiltro3 = yfiltro2[0:len(yfiltro2)-5000:1]                       #ELIMINAMOS LOS ULTIMOS CINCO MIL ELEMENTOS 
len(yfiltro3)
```

```python
#GRAFICO 
y = yfiltro3
x = np.array(np.arange(0,len(yfiltro3),1))                     #ocupando el nuevo len(yfiltro3)
plt.scatter(x,y)
plt.style.use('classic')
plt.grid(True)
plt.title('Grafico Datos 1  $aceleracion$ $v/s$ $ medicion $')
plt.ylabel('Acelarecion vertical $[m/s^2]$')
plt.xlabel('numero de medicion')
```

```python
#HISTOGRAMA
plt.hist(yfiltro3,bins=200)
plt.title('histograma Datos 1  $aceleracion$ $v/s$ $ medicion $')
```

```python
#AJUSTE POR MINIMOS CUADRADOS A UNA RECTA
coef = np.polyfit(x,y,1)                              #Obtengo los valores de los coeficientes
xx = np.linspace(np.min(x),np.max(x),100)            

yy = np.polyval(coef,xx)                              #EVALUAMOS LOS VALORES DE LOS COEFICIENTES EN XX1

plt.plot(xx,yy,color="red",label="ajuste")
plt.scatter(x,y,label="datos")
plt.legend()
plt.xlabel("")
plt.ylabel("")
plt.title("")
plt.grid(True)
plt.title('ajuste por minimos cuadrados Datos 1 ')
```

```python
#COEFICIENTE DE DETERMINACION

a1 ,a0 ,r ,_ , _ = linregress(x,y) 
St = np.sum((y-np.mean(y))**2)           #DIFERENCIA ENTRE LOS Y Y EL PROMEDIO DE DATOS 
res = (y-a0-a1*x)                        #RESIDUOS MINIMOS TRAS EL AJUSTE A LA FUNCION QUE SE TRABAJA
Sr = np.sum((res)**2)
r2 = ((St-Sr)/St)
print("r2 vale ", r2)
```

```python
#ANALISIS DE RESIDUOS A LA APROXIMACION

plt.scatter(x,res)
plt.xlabel("$x$")
plt.ylabel("residuos $\\epsilon$")
plt.title("analisis de residuos ")
```

```python
#VALOR MEDIO Y DESVIACION ESTANDAR
prom = np.mean(yfiltro3)
desviacion = np.std(yfiltro3)
print("Valor medio",prom)
print("Desviacion estandar",desviacion) 
```
