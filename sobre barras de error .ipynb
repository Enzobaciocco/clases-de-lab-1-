```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('classic')
#2_b) Gráfico V v/s t
plt.title('Grafico $V$ v/s $t$')##esto genera un tirulo
plt.ylabel('Diferencia de Potencial $[V]$')##esto genera una etiqueta para el eje y
plt.xlabel('Tiempo $[t]$')#etiqueta para x 
y = np.array([0.01, 1.43, 3.09, 3.63, 3.85, 4.19, 4.37, 4.45, 4.55, 4.61, 4.63, 4.63])#valores para cada y en modo arreglo
x = np.array([0.0, 0.3, 0.5, 0.8, 1.0, 1.3, 1.5, 1.8, 2.0, 2.3, 2.5, 2.8])#valores en x para el eje x 
plt.plot(x,y,marker='s',color='red',)
plt.errorbar(x,y, yerr= 0.005, xerr=0.05)#genera una barra de error en x e y con los valores asignados como largo de la barra 
plt.xlim(-0.5,3) #limites en el eje x 
plt.ylim(-0.5,5.5)#limnites en el eje y

```

```python
#2_d) Graficamos (v_inf-v)/v_inf considerando error zr = 0.001.
plt.title('Grafico $(V_{\infty}-v)/v_{\infty}$ v/s $t$')##genra titulo
plt.ylabel('Diferencia de Potencial $[V]$')#etiqueta en y
plt.xlabel('Tiempo $[t]$')#etiqueta en x 
y = np.array([0.01, 1.43, 3.09, 3.63, 3.85, 4.19, 4.37, 4.45, 4.55, 4.61, 4.63, 4.63])#datos de arreglo para y
x = np.array([0.0, 0.3, 0.5, 0.8, 1.0, 1.3, 1.5, 1.8, 2.0, 2.3, 2.5, 2.8])#datos en y como arreglo
z = (4.63-y)/4.63 #esto genera un nuevo arreglo con el valor obtenido del grafico 2b 
zr= 0.001 #valor de error para z como la mitad del valor de sensibilidad minima 

#grafica con ambos ejes normales.
plt.plot(x,z,marker='s',color='red',)#genera el grafico
plt.errorbar(x,z, yerr= zr, xerr=0.05)#genera las barras de error asociadas donde el error de y es zr y z un arreglo
plt.xlim(-0.2,3.0)#limites de x
plt.ylim(-0.1,1.1)#limites de y

```

```python

#Para la gráfica con semilogx:###esto hace que un eje tenga eje logaritmico en x 
plt.semilogx(x,z,marker='s',color='red',)#genera grafico con eje de logaritmos
plt.errorbar(x,z, yerr= zr, xerr=0.05)##en esta parte el error de y es segun zr que es un arreglo de datos generado
plt.xlim(1*10**(-0.4),3.0)#limites en x
plt.ylim(-0.1,0.4)#limites en y
```

```python

#Para la gráfica con semilogy:
plt.semilogy(x,z,marker='s',color='red',)##esto genera el grafico con escala log en y y solo en y 
plt.errorbar(x,z, yerr= zr, xerr=0.05)
plt.xlim(-0.5,3.0)
plt.ylim(1*10**(-2.6),1.4)
```

```python
#Para la gráfica con semilogy:
plt.loglog(x,z,marker='s',color='red',)
plt.errorbar(x,z, yerr= zr, xerr=0.05)
plt.xlim(1*10**(-0.6),3.0)
plt.ylim(1*10**(-2.6),1.4)
```
