```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
datos = np.genfromtxt('ej-ajuste-03.txt')
x = datos[:,0]
y = datos[:,1]
print(x)
len(x)
```

```python
plt.scatter(x,y)
plt.grid(True)
```

```python
m = 2#grado del polinomio
M = np.zeros((m+1,m+1))#generando una matriz de m+1 por m+1
#AHORA BVAMOS A HACER CICLOS PARA PODER MOVERNOS DE ARRRIBA A ABAJO Y DE ISUIERDA A DERECHA 
for p in range (m+1):#esto de arriba a abajo 
    for k in range (m+1):#esto se muve de isquierda a derecha 
        M[p,k] = np.sum(x**(p+k))
        
print(M)
print(np.sum(x**4))
```

```python
b = np.zeros(m+1)
for p in range(m+1):
    b[p] = np.sum(y*x**p)
print(b)
```

```python
np.linalg.solve(M,b)#esto resuelve el sistema a*x=b en nuestro caso M*x=b
```

```python
#ahora podemos asignar los valores resultos a su variable correspondida asi
a0,a1,a2 = np.linalg.solve(M,b)
```

```python
xx = np.linspace(np.min(x),np.max(x),100)
yy = a0 + a1*xx + a2*xx**2
plt.plot(xx,yy)
plt.scatter(x,y)
plt.grid(True)
```

```python
np.polyfit(x,y,2)#esto resuelve el sistema igual pero entrega los valores del mayor al menor grado 
```

```python
#ahora una mejor y mas rapida forma de calcular los y de la funcion con polyval
```

```python
coef = np.polyfit(x,y,14)#esto resuelve el sistema para la funcion de grado que uno quiere
xx = np.linspace(np.min(x),np.max(x),100)
#yy = a0 + a1*xx + a2*xx**2#esto se puede remplazar por 
yy = np.polyval(coef,xx)#evalua los y segun los a2 a1 a0 ##de esta forma uno no tiene que definir explicitamente la forma del polinomio 
plt.plot(xx,yy)
plt.scatter(x,y)
plt.grid(True)
```
