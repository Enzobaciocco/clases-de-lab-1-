```python
import numpy as np
import matplotlib.pyplot as plt
datos = np.genfromtxt('ejemplo-s11a.txt')
x = datos[:,0]
y = datos[:,1]
plt.scatter(x,y)
plt.grid(True)
```

```python
a1,a0 = np.polyfit(x,y,1)
print(a1,a0)
```

```python
xx = np.linspace(-1,11,1000)
yy = a0 + a1*xx 
plt.plot(xx,yy,color="red",label="ajuste")
plt.scatter(x,y,label="datos")
plt.legend()#esto hace ue se vean los label  
plt.grid(True)
```

```python
St = np.sum((y-np.mean(y))**2)#difenrecia entre los y el promedio 
res = (y-a0-a1*x)#residuos minimos tras el ajuste a la funcion ue uno este trabajando 
Sr = np.sum((res)**2)
r2 = ((St-Sr)/St)
print("r2 vale ", r2)
```
